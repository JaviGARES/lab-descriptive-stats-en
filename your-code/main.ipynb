#### Show visually the frequency distribution (histogram) of the employee dataset. In few words describe these histograms?

```python
# your answer here

num_cols = employee.select_dtypes(include="number").columns
cat_cols = employee.select_dtypes(include="object").columns
```

```python
# YOUR CODE HERE
for col in num_cols:
    employee[col].plot(kind= "hist")
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

Years distribution histogram:  

The distribution is uneven and discrete, with small peaks at 2 years and 8 years.
Most employees have around 2 years of experience, while a few have 78 years.
The data is right-skewed (more people with fewer years).

Salary histogram:

Salaries are spread out between 30 and 70 with a few common pay levels around 35k, 55-60k, and 70k

```python
# YOUR CODE HERE
for col in cat_cols:
    employee[col].value_counts().plot(kind="bar")
    plt.title(f"Frequency of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```

#### Show a visual summary of the data using boxplot. What Are the First and Third Quartiles? Comment your results.

* Hint : Quantiles vs Quartiles*
In Probability and Statistics, quantiles are cut points dividing the range of a probability distribution into continuous intervals with equal probabilities. When division is into four parts the values of the variate corresponding to 25%, 50% and 75% of the total distribution are called quartiles.

```python
# YOUR CODE HERE

employee[num_cols].boxplot(vert=False)  
plt.title("Horizontal Boxplot of Numeric Variables")
plt.xlabel("Values")  
plt.ylabel("Variables")
plt.show()


```

```python
# Year's barplot is right skewed distribution. Median is lower than mean.
# Salary barplot distribution is left skewed. Median is higher than mean.
```

```python
# print first quartile here
employee["Salary"].quantile(0.25)
```

```python
# print third quartile here
employee["Salary"].quantile(0.75)
```
